```python
# 으뜸 머신러닝 심화문제 3.2

import pandas as pd 

name = pd.Series(['A', 'B','C',
                         'D','E', 'F','G'])
horse_power = pd.Series([130,250,190,300,210,220,170])
weight = pd.Series([1.9,2.6,2.2,2.9,2.4,2.3,2.2])
efficiency = pd.Series([16.3, 10.2, 11.1, 7.1,12.1, 13.2,14.2])

df = pd.DataFrame({'name': name, 'horse power': horse_power, 'weight': weight, 'efficiency': efficiency})
df
```

```python
df_2 = df.set_index(keys=['name'], inplace=False, drop=True)
print(df_2)
```

```python
df['results'] = pd.DataFrame(df['horse power']*df['efficiency'])
df_2 = df.set_index(keys=['name'], inplace=False, drop=True)
print(df_2)

print('최대값')
print(df_2['results'].idxmax())
```
